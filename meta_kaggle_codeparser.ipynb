```python
import pandas as pd

kernels = pd.read_csv('/kaggle/input/superker/super_kernels2.csv')
```

```python
kernels_compact = kernels.drop(['TeamLeaderId', 'ScoreFirstSubmittedDate', 'PublicLeaderboardSubmissionId', 'PrivateLeaderboardSubmissionId', 'IsBenchmark', 'MedalAwardDate', 'SourceCompetitionId', 'AuthorUserId', 'WriteUpForumTopicId', 'Unnamed: 0', 'Id', 'Medal', 'LastSubmissionDate'], axis=1)
kernels_compact
```

```python
import os
import glob

def idToPath(id):
    f = str(id).zfill(10)
    prefix = '/kaggle/input/meta-kaggle-code/'+f[0:4]+'/'+f[4:7]+'/'+str(id)+'.*'
    g = glob.glob(prefix)
    if len(g) == 1:
        return g[0]
    return ""

def ftype(p):
    parts = os.path.splitext(p)
    if len(parts) != 2:
        return ""
    return parts[1]

def pathToId(p):
    parts = os.path.splitext(os.path.basename(p))
    if len(parts) != 2:
        return ""
    if parts[0] != "":
        return int(parts[0])
    return None

import codecs
import json
from pathlib import Path

def ipynbSource(path):
    f = codecs.open(path, 'r')
    source = f.read()
    
    jsource = json.loads(source)
    cells = []
    for cell in jsource['cells']:
        if cell['cell_type'] == 'code':
            cells.append(cell['source'])
    return cells

def sourceCode(path):
    t = ftype(path)
    if t == ".ipynb":
        return ipynbSource(path)
    return [Path(path).read_text()]

def sourceCodeById(id):
    p = idToPath(id)
    if p != "":
        return sourceCode(p)
    return None
```

```python
kernels_compact['code'] = kernels_compact['CurrentKernelVersionId'].apply(sourceCodeById)
```

```python
kernels_compact.to_csv('final_notebooks2.csv')
```
