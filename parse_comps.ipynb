```python
kernels_comps = pd.read_csv('KernelVersionCompetitionSources.csv')
kernels = pd.read_csv('KernelVersions.csv')
kernels_comps.tail()
```

```python
comps_ids = df_new_filtered['Id'].unique().tolist()
```

```python
kernels_comps_filt = kernels_comps[kernels_comps['SourceCompetitionId'].isin(comps_ids)]
kernels_comps_filt

kcf_list = kernels_comps_filt['KernelVersionId'].tolist()
```

```python
teams = pd.read_csv('Teams.csv')
teams_memberships = pd.read_csv('TeamMemberships.csv')
```

```python
teams_filt = teams[teams['CompetitionId'].isin(comps_ids)]
teams_filt = teams_filt.dropna(subset=['PublicLeaderboardRank'])
teams_filt
```

```python
teams_list = teams_filt['Id'].unique().tolist()
users = teams_memberships[teams_memberships['TeamId'].isin(teams_list)]
users = users.drop(['Id'], axis=1)
users
```

```python
Kernels = pd.read_csv('Kernels.csv')
```

```python
Kernels_new = Kernels.dropna(subset=['CurrentKernelVersionId'])
Kernels_new['CurrentKernelVersionId'] = Kernels_new['CurrentKernelVersionId'].astype('int')
Kernels_new_filt = Kernels_new[Kernels_new['CurrentKernelVersionId'].isin(kcf_list)]
kernels_new_filt =pd.merge(Kernels_new, kernels_comps.drop_duplicates(subset=['KernelVersionId']), left_on='CurrentKernelVersionId', right_on='KernelVersionId', how='left')
kernels_new_filt
```

```python

merged_df = pd.merge(teams_filt, users, left_on='Id', right_on='TeamId', how='inner')


kernels_go = kernels_new_filt[['AuthorUserId', 'CurrentKernelVersionId', 'SourceCompetitionId']]

final_df = pd.merge(merged_df, kernels_go, left_on=['UserId', 'CompetitionId'], right_on=['AuthorUserId', 'SourceCompetitionId'], how='inner')
```

```python
final_df
```

```python
final_df_new = final_df.drop_duplicates(subset=['CurrentKernelVersionId'])
final_df_new = final_df_new.drop_duplicates(subset=['Id'])
```

```python
competitions = final_df_new['CompetitionId'].unique().tolist()
filtered_by_rating = pd.DataFrame()

for comp in competitions:
    data = final_df_new[final_df_new['CompetitionId'] == comp]
    temp = data.sort_values('PublicLeaderboardRank', ascending=True).head(10000)
    filtered_by_rating = pd.concat([filtered_by_rating, temp], ignore_index=True)
```
